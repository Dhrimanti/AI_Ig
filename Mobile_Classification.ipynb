```python
x=df.drop(columns=['price_range'],axis=1)
y=df['price_range']
```

```python
y.describe()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
df
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
print(model.score(x_train,y_train))
```

```python
print(model.score(x_test,y_test))
```

```python
model.predict([[1500,0,1.5,1,4,1,8,0.1,180,2,5,500,750,2500,11,6,15,1,1,1]])
```

```python
from sklearn import tree
model2=tree.DecisionTreeClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
print(model2.score(x_test,y_test))
```

```python
model2.predict([[1500,0,1.5,1,4,1,8,0.1,180,2,5,500,750,2500,11,6,15,1,1,1]])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3=RandomForestClassifier(n_estimators=193)
```

```python
model3.fit(x_train,y_train)
```

```python
print(model3.score(x_test,y_test))
```
